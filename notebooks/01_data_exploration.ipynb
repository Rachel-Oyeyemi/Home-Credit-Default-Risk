# 01 – Data Exploration

In this notebook we explore the synthetic Home Credit sample dataset.  The goal is to understand distributions, relationships between variables and potential issues such as missing values or outliers.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

# Load the raw dataset
data_path = Path('..') / 'data' / 'raw' / 'home_credit_sample.csv'
df = pd.read_csv(data_path)

df.head()
```

```
# Summary statistics
df.describe()
```

```
# Plot distribution of numeric variables
num_cols = df.select_dtypes(include=['int64','float64']).columns
for col in num_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
```

```
# Correlation heatmap
import numpy as np
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
